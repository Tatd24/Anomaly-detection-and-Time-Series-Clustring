Question 1 : What is Dimensionality Reduction? Why is it important in machine
learning?

Ans- Many real-world datasets have hundreds or thousands of features. Processing such high-dimensional data is:
•	computationally expensive
•	difficult to visualize
•	prone to overfitting (the model learns noise instead of patterns)

Dimensionality Reduction transforms the data from a high-dimensional space into a lower-dimensional space.
This can be done by:
1. Feature Selection
Selecting a subset of the original features.
Examples:
•	Filter methods (correlation, Chi-square)
•	Wrapper methods
•	Embedded methods (Lasso, decision tree importance)
2. Feature Extraction
Creating new features by transforming the data.
Examples:
•	PCA (Principal Component Analysis)
•	t-SNE
•	LDA (Linear Discriminant Analysis)
•	Autoencoders

It is important in Machine Learning because-

1. Reduces Overfitting
High-dimensional data increases the chances of noise being learned.
Reducing dimensions ⇒ fewer parameters ⇒ better generalization.
2. Improves Model Performance
Less computational cost, faster training, and prediction.
3. Handles the “Curse of Dimensionality”
As dimensions increase, data becomes sparse and distances become meaningless.
Dimensionality reduction solves this problem.
4. Better Visualization
High-dimensional data (100s of features) can be projected into:
•	2D
•	3D
to visualize clusters or patterns.
5. Removes Multicollinearity
Highly correlated features can be merged (e.g., PCA), improving model stability.


Question 2: Name and briefly describe three common dimensionality reduction
techniques..

Ans- Here are three common dimensionality reduction techniques along with brief descriptions:
________________________________________
 1. Principal Component Analysis (PCA)
PCA is a linear dimensionality reduction technique that transforms the original features into a new set of uncorrelated variables called principal components.
•	These components capture the maximum variance in the data.
•	It is widely used for noise reduction, visualization, and speeding up ML models.
________________________________________
 2. t-Distributed Stochastic Neighbor Embedding (t-SNE)
t-SNE is a non-linear technique mainly used for visualizing high-dimensional data in 2D or 3D.
•	It preserves local structure (points close in high-dimensional space stay close in lower dimensions).
•	Great for visualizing clusters (e.g., images, embeddings).
•	Computation-heavy and usually not used for feature preprocessing in models.
________________________________________
 3. Linear Discriminant Analysis (LDA)
LDA is a supervised dimensionality reduction technique.
•	It reduces dimensions by maximizing the separation between classes.
•	Unlike PCA, which is unsupervised, LDA uses class labels.
•	Common in classification tasks like face recognition.



Question 3: What is clustering in unsupervised learning? Mention three popular
clustering algorithms.

Ans- Clustering is the process of grouping a set of data points in such a way that:
•	Points in the same cluster are more similar to each other
•	Points in different clusters are more dissimilar
It is used when the data has no labels, and the goal is to discover natural patterns or groupings in the dataset.
Clustering is widely used in:
•	Customer segmentation
•	Image segmentation
•	Anomaly detection
•	Document grouping
•	Biological data analysis

Three Popular Clustering Algorithms
1. K-Means Clustering
•	Divides data into K clusters
•	Each cluster has a centroid (mean point)
•	Iteratively assigns points to the nearest centroid
•	Best for well-separated, spherical clusters
________________________________________
2. Hierarchical Clustering
•	Builds a tree-like (dendrogram) structure of clusters
•	Can be:
o	Agglomerative (bottom-up) – merge small clusters
o	Divisive (top-down) – split big clusters
•	No need to pre-set the number of clusters
________________________________________
3. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)
•	Groups points based on density
•	Can find clusters of arbitrary shapes
•	Automatically identifies noise/outliers
•	Useful when clusters are not well-separated



Question 4: Explain the concept of anomaly detection and its significance.

Ans- Anomaly Detection (also called outlier detection) is the process of identifying data points or events that deviate significantly from the normal pattern in a dataset.
Anomalies are rare but often critical because they may represent:
•	Fraud
•	System failures
•	Security breaches
•	Unexpected behavior
•	Errors in data
In unsupervised learning, anomaly detection can be done without labeled examples, by learning what “normal” looks like and flagging deviations.

 Types of Anomalies
1.	Point Anomalies – A single data point that is unusual
o	Example: A sudden spike in credit card transactions
2.	Contextual Anomalies – An anomaly depending on context
o	Example: High temperature is normal in summer but abnormal in winter
3.	Collective Anomalies – A collection of data points is abnormal
o	Example: A series of unusual network requests indicating an attack

Anomaly detection is extremely important because:
1. Protects Against Fraud
Used in banking and finance to identify unusual transactions.
2. Enhances Security
Helps detect:
•	Network intrusions
•	Malware
•	Unauthorized access
3. Prevents System Failures
In manufacturing or IoT, anomalies can indicate:
•	Faulty machines
•	Sensor malfunctions
Early detection prevents costly damage.
4. Improves Data Quality
Flags errors, missing values, or corrupted records.
5. Detects Rare Events
Anomalies often represent events that are:
•	Rare
•	Valuable
•	Important to act upon

Question 5: List and briefly describe three types of anomaly detection techniques.

Ans-  Here are three major types of anomaly detection techniques, along with brief explanations:
________________________________________
1.	Statistical (Probability-Based) Methods
These methods assume that normal data follows a specific statistical distribution (e.g., Gaussian).
Anomalies are points that deviate significantly from this distribution.
Examples:
•	Z-score
•	Gaussian mixture models
•	Boxplot (IQR method)
2. Machine Learning–Based Methods
These use learning algorithms to model "normal" behavior and detect deviations.
Examples:
•	Isolation Forest
o	Isolates anomalies by random partitioning; anomalies are isolated faster.
•	One-Class SVM
o	Learns the boundary around normal points; anything outside is an anomaly.
•	Local Outlier Factor (LOF)
o	Detects anomalies based on local density deviations.
3. Deep Learning–Based Methods
Used for complex, high-dimensional data like images, sensor data, and sequences.
Examples:
•	Autoencoders
o	Train a neural network to reconstruct normal data; high reconstruction error ⇒ anomaly.
•	LSTM networks
o	Used for time-series anomaly detection.
•	Variational Autoencoders (VAE)


Question 6: What is time series analysis? Mention two key components of time series data

Ans-  Time Series Analysis is the process of analyzing data points that are collected or recorded over time at regular intervals (daily, monthly, yearly, etc.).
The goal is to understand underlying patterns, identify trends, and make future predictions.
Examples include:
•	Stock prices
•	Weather data
•	Sales over months
•	Heartbeat signals
Time series analysis helps understand how values change with respect to time.
Two Key Components of Time Series Data
1. Trend
•	The long-term upward or downward movement in the data.
•	Example: A gradually increasing monthly sales curve.
2. Seasonality
•	Regular, repeating patterns that occur at fixed intervals (daily, weekly, yearly).
•	Example: Higher ice-cream sales every summer.
Other Components (for your understanding)
Although you only asked for two, here are others often discussed:
•	Cyclic patterns (non-fixed repeating trends)
•	Noise or randomness (unpredictable variation)

Question 7: Describe the difference between seasonality and cyclic behavior in time series.

Ans-  Here is a clear and simple explanation of the difference between seasonality and cyclic behavior in time series:
Seasonality
Seasonality refers to regular and predictable patterns that repeat at fixed, known intervals.
Key Features:
•	Occurs at regular intervals (e.g., daily, weekly, monthly, yearly)
•	Pattern is repetitive and predictable
•	Caused by seasonal factors like weather, holidays, business cycles
Example:
•	Increased ice cream sales every summer
•	Higher electricity usage every evening
•	Retail sales spike every December
Cyclic Behavior
Cyclic behavior refers to patterns that rise and fall, but the duration and amplitude are NOT fixed.
Key Features:
•	Cycles do not repeat at fixed intervals
•	Length and intensity vary
•	Often influenced by economic or market forces
•	Less predictable than seasonality
Example:
•	Economic recession and recovery cycles
•	Long-term business cycles
•	Fluctuations in real estate markets

Question 8: Write Python code to perform K-means clustering on a sample dataset.
(Include your Python code and output in the code box below.)

```python
import numpy as np
from sklearn.cluster import KMeans
import pandas as pd

# Create a simple sample dataset
X = np.array([
    [1, 2], [1, 4], [1, 0],
    [10, 2], [10, 4], [10, 0]
])

# Perform K-means clustering
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(X)

# Prepare output
df = pd.DataFrame(X, columns=["Feature 1", "Feature 2"])
df["Cluster"] = kmeans.labels_

df
```

Question 9: What is inheritance in OOP? Provide a simple example in Python.

Ans-  Inheritance in Object-Oriented Programming (OOP) is a mechanism that allows one class (called the child or subclass) to acquire the properties and behaviors (attributes and methods) of another class (called the parent or superclass).

It helps with:

Code reusability

Organizing related classes

Extending functionality without rewriting code

```python
# Parent class
class Animal:
    def speak(self):
        return "This animal makes a sound."

# Child class inheriting from Animal
class Dog(Animal):
    def speak(self):
        return "The dog barks."

# Another child class
class Cat(Animal):
    def speak(self):
        return "The cat meows."


# Creating objects
dog = Dog()
cat = Cat()

print(dog.speak())  # Output: The dog barks.
print(cat.speak())  # Output: The cat meows.
```

Question 10: How can time series analysis be used for anomaly detection?


Ans-  Time series analysis is widely used for anomaly detection because it helps identify data points that do not follow expected temporal patterns.
Analysis Helps in Anomaly Detection
Time series data changes over time, so anomalies are detected by comparing current values with expected values based on historical patterns.
Anomaly detection in time series identifies:
•	Sudden spikes
•	Sudden drops
•	Unexpected changes in trends
•	Deviations from seasonal patterns
Methods for Time-Series Anomaly Detection
1. Statistical Forecasting (ARIMA, Exponential Smoothing)
•	Models like ARIMA predict future values.
•	If the actual value deviates significantly from the forecasted value → anomaly.
Example:
If predicted sales = 100 units but actual = 250 units → unusual spike.
2. Seasonal Decomposition
•	Time series is decomposed into:
o	Trend
o	Seasonality
o	Residual (noise)
•	Large residuals (errors) indicate anomalies.
________________________________________
3. Machine Learning Models
Models such as:
•	LSTM networks (for long-term patterns)
•	Isolation Forest
•	One-Class SVM
These learn normal temporal behavior and mark deviations as anomalies.
________________________________________
4. Moving Average & Thresholding
•	Calculate rolling mean and standard deviation.
•	Points that fall far outside normal range are flagged as outliers.
